### **Dataset 1 : Breast Cancer Wisconsin (Diagnostic)**

**Type:** 
CSV

**Method:**
The data was programmatically downloaded from UCI ML Repository

**Dataset variables:**

- 1. ID number
- 2. Diagnosis (M = malignant, B = benign)
- 3. ~32. 
    - Ten real-valued features are computed for each cell nucleus:

        - a) radius (mean of distances from center to points on the perimeter)
        - b) texture (standard deviation of gray-scale values)
        - c) perimeter
        - d) area
        - e) smoothness (local variation in radius lengths)
        - f) compactness (perimeter^2 / area - 1.0)
        - g) concavity (severity of concave portions of the contour)
        - h) concave points (number of concave portions of the contour)
        - i) symmetry
        - j) fractal dimension ("coastline approximation" - 1)

- The mean, standard error and "worst" or largest (mean of the three
largest values) of these features were computed for each image,
resulting in 30 features. For instance, field 3 is Mean Radius, field
13 is Radius SE, field 23 is Worst Radius.

- All feature values are recoded with four significant digits.

- Missing attribute values: none

- Class distribution: 357 benign, 212 malignant

**breast_cancer_wisconsin_diagnostic**

```python
import requests
import csv
from io import StringIO

# Fetch data from the UCI ML Repository (as shown in the previous answer)
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data"

response = requests.get(url)

response.raise_for_status()
```

```python
if response.status_code == 200:
    # Parse the CSV data
    data = response.text
    data_io = StringIO(data)  
    csv_reader = csv.reader(data_io)
    
    # Convert the parsed CSV data to a Pandas DataFrame
    df1 = pd.DataFrame(csv_reader)

else:
    print(f"Failed to retrieve data. Status code: {response.status_code}")
```

#### Column Names

```python
columns = ['ID', 'Diagnosis','radius1', 'texture1', 'perimeter1', 'area1', 'smoothness1',
       'compactness1', 'concavity1', 'concave_points1', 'symmetry1',
       'fractal_dimension1', 'radius2', 'texture2', 'perimeter2', 'area2',
       'smoothness2', 'compactness2', 'concavity2', 'concave_points2',
       'symmetry2', 'fractal_dimension2', 'radius3', 'texture3', 'perimeter3',
       'area3', 'smoothness3', 'compactness3', 'concavity3', 'concave_points3',
       'symmetry3', 'fractal_dimension3']
```

```python
# Apply the cleaning strategy
column_to_change = ['radius1', 'texture1', 'perimeter1', 'area1', 'smoothness1',
       'compactness1', 'concavity1', 'concave_points1', 'symmetry1',
       'fractal_dimension1', 'radius2', 'texture2', 'perimeter2', 'area2',
       'smoothness2', 'compactness2', 'concavity2', 'concave_points2',
       'symmetry2', 'fractal_dimension2', 'radius3', 'texture3', 'perimeter3',
       'area3', 'smoothness3', 'compactness3', 'concavity3', 'concave_points3',
       'symmetry3', 'fractal_dimension3']

for i in column_to_change:
    df1_mod[i]=df1_mod[i].astype('float')
```

### Quality Issue on Dataset 2: Completeness issue, Uniqueness issue

**Completeness issue**

```python
df2_mod=df2_mod.drop(['Unnamed: 3'], axis=1)
df2_mod.head()
```

**Uniqueness issue**

```python
# Remove duplicated data
df2_mod = df2_mod[~df2_mod.duplicated()==True]
df2_mod.duplicated().sum()
```

### **Tidiness Issue 1: Clean column names of Dataset 1**


Since **Dataset 1 is a tidy data**, the only part that we may adjust to make the dataset better is the naming for variables. For the number at the end of variables (ex: radius"1", radius"2", radius"3"), these stands for the **"1:Mean"**, **"2:Standard error"**, and **"3:Worst case"**. For example, radius"1", radius"2", radius"3" represent "Mean radius", "Radius SE", and "Worst radius".

```python
new_col = list(df1_mod.columns)

for i in range(len(new_col)):
    new_col[i]=new_col[i].replace('1','_mean')
    new_col[i]=new_col[i].replace('2','_se')
    new_col[i]=new_col[i].replace('3','_worst')

df1_mod.columns=new_col
df1_mod.head()
```

### **Tidiness Issue 2: Separate multiple variables stored in "Grade" column**

Variable "Grade" contains the grade value for tumor and descrption of differentiated level. Here, we will separate the information into 2 variables: **"Grade"** and **"Differentiated_level"**

```python
df2.Grade.value_counts()
```

```python
#FILL IN - Apply the cleaning strategy
grade_list = list(df2_mod['Grade'])
```

```python
a = []
b = []

for i in range(len(grade_list)):
    grade_list[i]= grade_list[i].replace('; anaplastic','')
    a.append(grade_list[i].split(';')[0]) 
    b.append(grade_list[i].split(';')[1]) 

df2_mod['Differentiated_level']=a
df2_mod['Grade']=b
```
